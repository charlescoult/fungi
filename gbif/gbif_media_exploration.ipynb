```python
media = pd.read_csv(
    'fungi/multimedia.txt',
    delimiter='\t',
)
media.shape
```

```python
media.columns.tolist()
```

```python
# Number of images per observation
media_vc = media['gbifID'].value_counts()
```

```python
media_vc.sort_values().hist( bins = 200 )
```

```python
# plot counts on a 1D line graph
import matplotlib.pyplot as pp
pp.plot(media_vc, len(media_vc) * [1], "x")  # is there a way to do a 1-dimensional plot within pandas?
```
