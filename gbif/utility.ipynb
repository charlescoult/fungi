```python
import pandas as pd

# Get columns

def get_col_all_value_unique( df ):
    all_unique_cols = []
    for col in df.columns:
        if ( len(df[col].value_counts()) == len(df.index) ):
            all_unique_cols.append( col )
    return all_unique_cols

def get_col_all_nan( df ):
    nan_cols = []
    for col in df.columns:
        if ( df[col].isnull().all() ):
            nan_cols.append(col)
    return nan_cols

def get_col_single_value( df ):
    sv_cols = []
    for col in df.columns:
        if( len(df[col].value_counts(dropna=False) ) <= 1 ):
            sv_cols.append(col)
    return sv_cols

# Remove columns

def remove_col( df, cols ):
    return df.drop( cols, axis=1 )

def remove_col_nan( df ):
    return remove_col( df, get_col_nan( df ) )

def remove_col_single_value( df ):
    return remove_col( df, get_col_single_value( df ) )

def remove_col_by_regex( df, regexes ):
    print(regexes)
    if (not isinstance(regexes, list)):
        regexes = [ regexes ]
    print(regexes)
    drop_cols = []
    for regex in regexes:
        print(df.filter(regex=regex).columns.tolist())
        drop_cols = drop_cols + df.filter(regex=regex).columns.tolist()
        print(drop_cols)
        print(df.filter(regex=regex).columns.tolist())
    return remove_col( df, drop_cols )

# Print info

def print_value_counts( df, columns = None ):
    if (columns == None):
        columns = df.columns
    for col in columns:
        print(col)
        print(df[col].value_counts())
        print()

def print_hasnans(
    df,
    columns = None,
    only_true = False,
):
    if (columns == None):
        columns = df.columns
    for col in columns:
        if ( df[col].hasnans or not only_true ):
            print(f'{col}: {df[col].hasnans}')
```
