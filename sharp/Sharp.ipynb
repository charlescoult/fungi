```python
high_page_count_df = df[ df['Number of Completed Pages'] > 100 ] \
.sort_values( by='Number of Completed Pages', ascending=False )
```

```python
high_page_count_df[''
```

```python
high_page_count_df['month'] = high_page_count_df['start_time'].dt.month
```

```python
high_page_count_df['year'] = high_page_count_df['start_time'].dt.year
```

```python
high_page_count_df['day'] = high_page_count_df['start_time'].dt.day
```

```python
high_page_count_df[['month', 'day', 'year', 'Number of Completed Pages']].head()
```

```python
high_page_count_df.groupby(['month', 'day'])['Number of Completed Pages'].sum().plot(kind='bar')
```

```python
high_page_count_df['time'] = high_page_count_df['start_time'].dt.time
```
