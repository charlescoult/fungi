```python
import pandas as pd
df = pd.read_csv('sharp.csv')
```

```python
df.shape
```

```python
def remove_col_nan( df ):
    nan_cols = []
    for col in df.columns:
        if ( df[col].isnull().all() ):
            nan_cols.append(col)
    print(f'Removing: {nan_cols}')
    return df.drop( nan_cols, axis=1 )

df = remove_col_nan( df )
df.shape
```

```python
def remove_col_single_value( df ):
    sv_cols = []
    for col in df.columns:
        if( (df[col] == df[col][0]).all() ):
            sv_cols.append(col)
    print(f'Removing: {sv_cols}')
    return df.drop(sv_cols, axis=1)
df = remove_col_single_value( df )
df.shape
```

```python
df.columns.tolist()
```

```python
filter_cols = df.filter(regex=':').columns.tolist()
```

```python
def print_value_counts( df, columns = None ):
    if (columns == None):
        columns = df.columns
    for col in columns:
        print(col)
        print(df[col].value_counts())
        print()
print_value_counts(df, filter_cols)
```

```python
# still not getting all single value columns
def remove_col_single_value( df ):
    sv_cols = []
    for col in df.columns:
        if( len(df[col].value_counts()) <= 1 ):
            sv_cols.append(col)
    print(f'Removing: {sv_cols}')
    return df.drop(sv_cols, axis=1)
df = remove_col_single_value( df )
df.shape
```

```python
def print_hasnans(
    df,
    columns = None,
    only_true = False,
):
    if (columns == None):
        columns = df.columns
    for col in columns:
        if ( df[col].hasnans or not only_true ):
            print(f'{col}: {df[col].hasnans}')

print_hasnans( df, [
    'Starting Date & Time',
    'Completing Date & Time',
] )
```

```python
def get_cols_all_unique( df ):
    all_unique_cols = []
    for col in df.columns:
        if ( len(df[col].value_counts()) == len(df.index) ):
            all_unique_cols.append( col )
    return all_unique_cols

all_unique = get_cols_all_unique( df )
print(all_unique)
```
